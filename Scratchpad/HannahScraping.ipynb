```python
import requests
import json
from bs4 import BeautifulSoup
import pprint
```

```python
url = "http://www.boxofficemojo.com/intl/china/yearly/?yr=2015&p=.htm"
```

```python
import html5lib
import bs4.builder._html5lib
import re
```

```python
def extract_location_data(url):
    response = requests.get(url)
    page_html = response.text
    #print page_html
    page_as_string = str(page_html)
    
    start = re.search('<h1>China Yearly Box Office</h1>', page_as_string).start()
    print "start 1", start
    
    temp_end = re.search('<font face="Verdana" size="2">Rows: <font size="3"><b><font size="4">#1&#150;100</font>', page_as_string).end()
    temp_end1 = re.search('<font face="Verdana" size="2">Rows: <font size="3"><b><font size="4">#1&#150;100</font>', page_as_string[temp_end:]).start()
    end = temp_end + temp_end1
    print "end 1", end
    
    
    page_html_subset = page_as_string[start:end]
    soup = BeautifulSoup(page_html_subset)
    #print soup
    # Retrieve all of the script tags
    td_collection = soup('td')
    for td in td_collection:
        print td
```
